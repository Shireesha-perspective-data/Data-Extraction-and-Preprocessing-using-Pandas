```python
'''*  function called to_datetime from the pandas i.e we are calling "to_datetime" function from the pandas
to_datetime fun or method from the pandas the born object is converted to datetime formate.
born object type is converted to datetime formate
* born object type is converted to to_datetime formate
'''
pd.to_datetime(df['Born'],format ="%Y-%m-%d") #here  we use datetime method to connected to pandas ,born colum is a object type ,format parameter is used  
df['born_datetime']=pd.to_datetime(df['Born'],format ="%Y-%m-%d") #
df
```

```python
df['died_datetime']=pd.to_datetime(df['Died'],format="%Y-%m-%d")  # Died object type is converted to datetime formate
df
```

```python
df.info()
```

```python
'''
 substracted  from the died_datetime to born_datetime .then we are how many days are we are live
 
'''
df['day_datetime']=df['died_datetime']-df['born_datetime'] # the  died_datetime is substracted to born_datetime 
```
